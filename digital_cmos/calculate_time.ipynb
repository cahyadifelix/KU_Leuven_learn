```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
delay_htl = np.array([11.424, 20.662, 30.378])
delay_lth = np.array([16.45, 25.77, 35.22])
h_array = np.array([0,1,2])
```

```python
lin_htl = linregress(h_array, y = delay_htl)
lin_lth = linregress(h_array, y = delay_lth)

```

```python
Fig, ax = plt.subplots(nrows= 1, ncols= 1, figsize = (9,5))
ax.scatter(h_array, delay_htl, label = "high-to-low")
ax.scatter(h_array, delay_lth, label = "low-to-high", color = "orange")
ax.axline((0,lin_htl.intercept), slope = lin_htl.slope, label = f"high-to-low fit, $r = {lin_htl.rvalue:1f}$")
ax.axline((0,lin_lth.intercept), slope = lin_lth.slope, label = f"low-to-high fit, $r = {lin_lth.rvalue:1f}$", color = "orange")
ax.set_xlabel("$h$", fontsize = 16)
ax.set_ylabel("delay (ps)", fontsize = 16)
ax.set_xlim([-0.02,2.02])
ax.set_ylim([-0.1, 40])
ax.legend()
ax.grid(True)
```
